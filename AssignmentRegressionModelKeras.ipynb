Regression Model in Keras Assignment

```python
import pandas as pd
import numpy as np

concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength']
n_cols=predictors.shape[1]
```

A. Build a baseline model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
#part 1
X_train, X_test, y_train, y_test=train_test_split(predictors, target, test_size=0.30) 
```

```python
#part 2
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
    
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=2)
```

```python
from sklearn.metrics import mean_squared_error
#part 3
y_pred=model.predict(X_test)
mean_squared_error(y_pred,y_test)
```

```python
#part 4
n=50
mean_sqr_errs=[]
for i in range(n):
    X_train_i, X_test_i, y_train_i, y_test_i=train_test_split(predictors, target, test_size=0.30)
    model_i = Sequential()
    model_i.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model_i.add(Dense(10, activation='relu'))
    model_i.add(Dense(1))
    model_i.compile(optimizer='adam', loss='mean_squared_error')
    model_i.fit(X_train_i, y_train_i, validation_split=0.3, epochs=50, verbose=2)
    mean_sqr_err_i=mean_squared_error(model_i.predict(X_test_i),y_test_i)
    mean_sqr_errs.append(mean_sqr_err_i)
mean_sqr_errs
```

```python
mean=np.mean(mean_sqr_errs)
print('The mean of the mean square errors is {}'.format(mean))
```

```python
std=np.std(mean_sqr_errs)
print('The standard deviation of the mean square errors is {}'.format(std))
```

We can observe that the standard deviation is huge which suggests that our model training isn't working that well

Part B

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std() #normalizing the data
n=10 #The description says 50 but the run in part A took so long I'm only doing 10 this time
mean_sqr_errs_B=[]
for i in range(n):
    X_train_i, X_test_i, y_train_i, y_test_i=train_test_split(predictors_norm, target, test_size=0.30)
    model_i = Sequential()
    model_i.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model_i.add(Dense(10, activation='relu'))
    model_i.add(Dense(1))
    model_i.compile(optimizer='adam', loss='mean_squared_error')
    model_i.fit(X_train_i, y_train_i, validation_split=0.3, epochs=50, verbose=2)
    mean_sqr_err_i=mean_squared_error(model_i.predict(X_test_i),y_test_i)
    mean_sqr_errs_B.append(mean_sqr_err_i)
mean_sqr_errs_B
```

```python
mean=np.mean(mean_sqr_errs_B)
print('The mean of the mean square errors is {}'.format(mean))
std=np.std(mean_sqr_errs_B)
print('The standard deviation of the mean square errors is {}'.format(std))
```

Now the standard deviation is a lot less so the training worked a lot better this time

Part C

```python
n=7 #doubling the epochs will double the run time so I'll limit to 7 this time
mean_sqr_errs_C=[]
for i in range(n):
    X_train_i, X_test_i, y_train_i, y_test_i=train_test_split(predictors_norm, target, test_size=0.30)
    model_i = Sequential()
    model_i.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model_i.add(Dense(10, activation='relu'))
    model_i.add(Dense(1))
    model_i.compile(optimizer='adam', loss='mean_squared_error')
    model_i.fit(X_train_i, y_train_i, validation_split=0.3, epochs=100, verbose=2)
    mean_sqr_err_i=mean_squared_error(model_i.predict(X_test_i),y_test_i)
    mean_sqr_errs_C.append(mean_sqr_err_i)
mean_sqr_errs_C
```

```python
mean=np.mean(mean_sqr_errs_C)
print('The mean of the mean square errors is {}'.format(mean))
std=np.std(mean_sqr_errs_C)
print('The standard deviation of the mean square errors is {}'.format(std))
```

The mean is now a lot lower than in part B so doubling the epochs got us much closer to properly finished training

Part D

```python
n=5 #Limiting further to 5 this time
mean_sqr_errs_D=[]
for i in range(n):
    X_train_i, X_test_i, y_train_i, y_test_i=train_test_split(predictors_norm, target, test_size=0.30)
    model_i = Sequential()
    model_i.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model_i.add(Dense(10, activation='relu'))
    model_i.add(Dense(10, activation='relu'))
    model_i.add(Dense(10, activation='relu'))
    model_i.add(Dense(1))
    model_i.compile(optimizer='adam', loss='mean_squared_error')
    model_i.fit(X_train_i, y_train_i, validation_split=0.3, epochs=50, verbose=2)
    mean_sqr_err_i=mean_squared_error(model_i.predict(X_test_i),y_test_i)
    mean_sqr_errs_D.append(mean_sqr_err_i)
mean_sqr_errs_D
```

```python
mean=np.mean(mean_sqr_errs_D)
print('The mean of the mean square errors is {}'.format(mean))
std=np.std(mean_sqr_errs_D)
print('The standard deviation of the mean square errors is {}'.format(std))
```

Mean square errors increased slightly which is most likely a sign that the more complicated neural network is overfitting slightly - or at least that in this case it wasn't worth it to use a deeper neural network
